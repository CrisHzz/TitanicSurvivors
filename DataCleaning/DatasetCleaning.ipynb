# Clean a NaN values

*The dataset have too many values in NaN and is neccesary delete him o replace using a some stadticic tecnical*

## if the person has the ship's box in NaN concludes that he/she did not survive?

```python
boat_survived=originalDataset[['survived','boat']]

boats_array=np.array(boat_survived['boat'])

boats_array = boats_array[~pd.isnull(boats_array)]

boat_survived.head(10)
```

*Exists three cases for the survivors , survived with a number boat, survived with a letter boat or survived but dont have a register with a boat*

```python
#First case , survivor but dont have a register 
def combination_survivor_boat_nan(survivor):
    cont = 0
    for _, row in boat_survived.iterrows():
        if row['survived'] == survivor and pd.isna(row['boat']):
            cont += 1

    return f'Number of combinations: {cont}'
        
print(combination_survivor_boat_nan(0)) #Dont survived

print(combination_survivor_boat_nan(1)) #Survived
```

```python
#Second case , survivor and the boat is a letter 

def combination_survivor_boat_letter(survivor):
    possible_letter_boats=['A','B','C','D']

    cont = 0

    for _, row in boat_survived.iterrows():
        if row['survived'] == survivor and row['boat'] in possible_letter_boats:
            cont += 1

    return f'Number of combinations: {cont}'


print(combination_survivor_boat_letter(0))

print(combination_survivor_boat_letter(1))
```

```python
#Third case, survivor and the boat is a number

def combination_survivor_boat_number(survivor):
    cont = 0
    for _, row in boat_survived.iterrows():
        if row['survived'] == survivor and isinstance(row['boat'], (int, float)):
            cont += 1
    return f'Number of combinations: {cont}'


print(combination_survivor_boat_number(0))

print(combination_survivor_boat_number(1))
```

*With all this we can conclude that the data can be influenced by people who have values in their box boat but did not survive, proceed to delete them and clean the data.*

```python
# Filter out rows where survived is 0 and boat is not NaN
condition = (originalDataset['survived'] == 0) & (originalDataset['boat'].notna())

# Remove those rows from the DataFrame
originalDataset = originalDataset.drop(originalDataset[condition].index)

originalDataset
```

```python
def comprobate_Nan_column(column:str):
    return originalDataset[column].isna().sum()

originalDataset.drop(columns='fare_bin',inplace=True)
```

```python
print(f"Number of NaN values in the 'cabin' column: {comprobate_Nan_column('cabin')}")

print(f"Number of NaN values in the 'Name' column: {comprobate_Nan_column('name')}")

print(f"Number of NaN values in the 'sex' column: {comprobate_Nan_column('sex')}")

print(f"Number of NaN values in the 'age' column: {comprobate_Nan_column('age')}")

print(f"Number of NaN values in the 'embarked' column: {comprobate_Nan_column('embarked')}")

print(f"Number of NaN values in the 'boat' column: {comprobate_Nan_column('boat')}")

print(f"Number of NaN values in the 'ticket' column: {comprobate_Nan_column('ticket')}")

print(f"Number of NaN values in the 'fare' column: {comprobate_Nan_column('fare')}")
```

*We need to clean the data first for embarked,fare with a simple delete NaN*

```python
originalDataset.dropna(subset=['fare'], inplace=True)

print(f"Number of NaN values in the 'fare' column: {comprobate_Nan_column('fare')}")


originalDataset.dropna(subset=['embarked'],inplace=True)

print(f"Number of NaN values in the 'embarked' column: {comprobate_Nan_column('embarked')}")

```

```python
survivor_counts = (originalDataset['survived'].value_counts())

#No survived
print(survivor_counts[0])

#Survived
print(survivor_counts[1])
```

*next like a age is numeric parameter , we will use the mean for fill the NaN places in it row*



```python

modes = originalDataset['age'].mode().head(3)

originalDataset['age'] = originalDataset['age'].apply(lambda x: np.random.choice(modes) if pd.isna(x) else x)

originalDataset
```

```python

# Generate a sequence of numbers in the range 1000 to 2000
numbers_to_assign = np.arange(1000, 1799)

# Get the indexes of the unassigned rows in 'boat'.
nan_indices = originalDataset[originalDataset['boat'].isna()].index

num_to_fill = min(len(nan_indices), len(numbers_to_assign))

# Assign the numbers to the corresponding positions
originalDataset.loc[nan_indices[:num_to_fill], 'boat'] = numbers_to_assign[:num_to_fill]

```

```python
originalDataset
```

```python
# Remove rows with NaN values in the 'cabin' column
originalDataset.dropna(subset=['cabin'], inplace=True)

#assign a generic cabin based on the passenger class
def imputar_cabin(row):
    if pd.isna(row['cabin']):
        if row['pclass'] == 1:
            return 'C1'
        elif row['pclass'] == 2:
            return 'C2'
        else:
            return 'C3'
    else:
        return row['cabin']

originalDataset['cabin'] = originalDataset.apply(imputar_cabin, axis=1)
```

```python
#Round the age with 2 decimals

originalDataset['age']=np.round(originalDataset['age'],2)

originalDataset


```
