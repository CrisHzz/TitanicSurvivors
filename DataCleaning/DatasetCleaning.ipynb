# Data Cleaning

*The original dataset contains too many errors, such as empty data, large numbers and unnecessary data. Before developing the project, we need to clean and repair the data to reduce the errors to future and optimize the machine learning model*

```python
import pandas as pd

path:str ='../Media/titanicOriginal.xls'

originalDataset=pd.read_excel(path)

originalDataset.head(30)
```

## Drop some columns that are unnecessary at first glance.

*This dataset contains a unnecessary data and is fundamental drop a some columns that dont be useful for the model, but another columns have to create a analysis to determinated if we can drop it*

```python
originalDataset=originalDataset.drop(columns=['home.dest','body'])

originalDataset
```

```python

correlation_sibsp = originalDataset['sibsp'].corr(originalDataset['survived'])
correlation_parch = originalDataset['parch'].corr(originalDataset['survived'])

print(f'Correlation between SibSp and Survived: {correlation_sibsp}')
print(f'Correlation between Parch and Survived: {correlation_parch}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graphic bar  SibSp
sns.barplot(x='sibsp', y='survived', data=originalDataset)
plt.title('Survival as a function of the number of siblings/spouses on board')
plt.show()

# Graphic bar  parch
sns.barplot(x='parch', y='survived', data=originalDataset)
plt.title('Survival as a function of the number of parents/children on board')
plt.show()
```

*Now we can see that having parents or siblings on the titanic is not a determining factor in surviving this disaster.

```python
originalDataset=originalDataset.drop(columns=['sibsp','parch'])

originalDataset
```
